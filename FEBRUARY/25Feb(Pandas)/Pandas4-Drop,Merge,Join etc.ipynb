```python
data1 = pd.DataFrame({     #giving custom index
    "key1": [1,2,3,4,5,6],
    "key2": [4,5,6,7,8,9],
    "key3": [3,4,5,6,7,8],
},
    index = ["a","b","c","d","e","f"]
)
data1
```

```python
data1.join(data2) #Works similar to merge works and default how = "left"
```

```python
data1.join(data2,how="inner") #Works similar to merge
```

```python
data1.join(data2,how="outer") #Works similar to merge
```

```python
data1.join(data2,how="cross") #Works on index
```
