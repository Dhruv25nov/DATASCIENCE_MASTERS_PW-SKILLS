```python
x = ["a","b","c","d","e"] #Categorical data
y = np.random.rand(5) #Numerical
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Vertical Bar Chart")
plt.bar(x,y)
plt.show()
```

```python
x = ["a","b","c","d","e"] #Categorical data
y = np.random.rand(5) #Numerical

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Horizontal Bar Chart")

plt.barh(x,y)
plt.show()
```
